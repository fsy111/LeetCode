# 5. Longest Palindromic Substring
## Description
Given a string s, find the longest palindromic substring in s. You may assume that the maximum length of s is 1000.

```python
class Solution:
    
    def LongestCommonSubstring(self, s, t):
        m = len(s)
        n = len(t)
        l = [[0] * n] * m
        longest = 0
        ret = ''
        for i in range(m):
            for j in range(n):
                if s[i] == t[j]:
                    if i == 0 or j == 0:  l[i][j] = 1
                    else: l[i][j] = l[i-1][j-1] + 1
                if l[i][j] > longest:
                    longest = l[i][j]
                    ret = s[i - longest + 1:i+1]
        return ret    
    
    
    def longestPalindrome(self, s: str) -> str:
    ### Method 1: Dynamic Programming
    ### p[i][j] = (s[i:j+1] is a palindrome)
    ### Then p[i][j] = p[i+1][j-1] and s[i] == s[j]
    ### The base cases are p[i][i] = True, p[i][i+1] = (s[i] == s[i+1])
    
        def dp(s):
            n = len(s)
#             if not s: return s
#             p = [[False] * n for _ in range(n)]
#             p[0][0] = True
#             longest = 1
#             ret = s[0:1]
#             for i in range(1, n):
#                 p[i][i] = True
#                 if s[i-1] == s[i]:
#                     p[i-1][i] = True
#                     longest = 2
#                     ret = s[i-1:i+1]
            
#             for i in range(n-2, -1, -1):
#                 for j in range(i+2, n):
#                     p[i][j] = p[i+1][j-1] and s[i] == s[j]
#                     if p[i][j] and (j - i + 1)>longest:
#                         longest = j-i+1
#                         ret = s[i:j+1]
            
        ### Optimize the space usage
        ### Time: O(n^2); Space: O(n)
            p = [False] * n
            longest = 0
            ret = ''
            for i in range(n-1, -1, -1):
                for j in range(n-1, i-1, -1):
                    if j==i: 
                        p[j] = True
                    elif j==i+1: 
                        p[j] = (s[i] == s[j])
                    else:
                        p[j] = p[j-1] and s[i] == s[j]
                    if p[j] and (j - i + 1)>longest:
                        longest = j-i+1
                        ret = s[i:j+1]
            return ret
        
        
        ### Method 2: expand palindrome from center
        ### Time: O(n^2); Space: O(1)
        def expand(s):
            def expandcenter(l,r,s):
                while l>=0 and r<n and s[l] == s[r]:
                    l -= 1
                    r += 1
                return r-l-1
            
            n = len(s)
            start, end = 0, 0
            for i in range(n):
                len_odd = expandcenter(i,i,s)
                len_even = expandcenter(i,i+1,s)
                length = max(len_odd, len_even)
                if length > end - start + 1:
                    start = i - (length-1) // 2
                    end = i + length // 2
            return s[start:end+1]
            
        
        return expand(s)

Solution().longestPalindrome('cababa')
```

```python
### METHOD 3: Manacher algorithm
### Time: O(n); Space: O(n)
def AddBoundary(s:str) -> str:
    if not s: return '||'
    return '|' + '|'.join(list(s)) + '|'

def RemoveBoundary(s:str) -> str:
    if len(s)<3 : return ''
    return s[1::2]

def Manacher(s:str):
    smod = AddBoundary(s)
    c, r = 0, 0
    length = len(smod)
    p = [0] * length
    for i in range(1, length):
        ## If the center of current substring is within the furthest parlindrome, check the mirror
        ## else, set start and end points and expand.
        if i < r:   
            mirror = c*2-i
            ## If the mirror parlindrome substring does not exceed the left boundary, 
            ## then length of right parlindrome = the length of its mirror;
            ## else, the length of right parlindrome >= r-i and we are about to expand.
            p[i] = min(p[mirror], r-i)
            
        ## Set the start and end of expansion.
        ## Note that if p[i]==p[mirror], then p[mirror] < r-i, 
        ## then p[mirror] < r-i-1 because smod[r] must be '|' so smod[r-1] is a letter.
        ## In this case, the next while loop will be skipped automatically,
        ## because smod[i+p[i]+1] and smod[2*i-end] are different letters, 
        ## by our definition of p[mirror].
        end = i+p[i]+1
        start = 2*i-end
        
        ## Expand
        while start>=0 and end<length and smod[start]==smod[end]:
            start -= 1
            end += 1
            p[i] += 1
            
        ## If current parlindrome exceed the right boundary of the furthest parlindrome,
        ## then update the center and the right boundary of the furthest parlindrome.
        if i+p[i] > r:
            c, r = i, i+p[i]
        
    maxlen = 0   
    for i in range(1, length):
        if p[i] > maxlen:
            c, maxlen = i, p[i]
    return RemoveBoundary(smod[c-maxlen:c+maxlen+1])

Manacher('dcacbcace')
        
```
