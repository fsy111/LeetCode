# 450. Delete Node in a BST
## Description
Given a root node reference of a BST and a key, delete the node with the given key in the BST. Return the root node reference (possibly updated) of the BST.

Basically, the deletion can be divided into two stages:

- Search for a node to remove.
- If the node is found, delete the node.

```python
### RECURSION

class Solution:
    def RightMost(self, node:TreeNode) -> TreeNode:
        parent = None
        while node.right:
            parent = node
            node = node.right
        return node, parent

    def LeftMost(self, node:TreeNode) -> TreeNode:
        parent = None
        while node.left:
            parent = node
            node = node.left
        return node, parent
    
    def deleteNode(self, root: TreeNode, key: int) -> TreeNode:
        
        if not root: return None
        if root.val > key:
            root.left = self.deleteNode(root.left, key)
        elif root.val < key:
            root.right = self.deleteNode(root.right, key)
        else:
            if root.left:
                node, parent = self.RightMost(root.left)
                if not parent:
                    node.right = root.right
                else:
                    parent.right = node.left
                    node.right = root.right
                    node.left = root.left
            elif root.right:
                node, parent = self.LeftMost(root.right)
                if not parent:
                    node.left = root.left
                else:
                    parent.left = node.right
                    node.left = root.left
                    node.right = root.right
            else: node = None
            root = node            
        return root    
```

```python
### ITERATION

class Solution:
    
    def deleteNode(self, root: TreeNode, key: int) -> TreeNode:
        
        def RightMost(node:TreeNode) -> TreeNode:
            parent = None
            while node.right:
                parent = node
                node = node.right
            return node, parent
        
        def LeftMost(node:TreeNode) -> TreeNode:
            parent = None
            while node.left:
                parent = node
                node = node.left
            return node, parent
        
        cur = root
        p = None
        flag = 0
        while cur:
            if key == cur.val:
                break
            else:
                p = cur
                if key < cur.val:  cur = cur.left
                else:  cur = cur.right
        if cur:
            if cur.left:
                node, parent = RightMost(cur.left)
                if not parent:
                    node.right = cur.right
                else:
                    parent.right = node.left
                    node.right = cur.right
                    node.left = cur.left
            elif cur.right:
                node, parent = LeftMost(cur.right)
                if not parent:
                    node.left = cur.left
                else:
                    parent.left = node.right
                    node.left = cur.left
                    node.right = cur.right
            else: node = None
            if not p:
                root = node
            else:
                if key < p.val:
                    p.left = node
                else:
                    p.right = node
        return root
```
