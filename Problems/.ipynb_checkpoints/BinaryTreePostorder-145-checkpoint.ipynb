```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def postorderTraversal(self, root: TreeNode) -> List[int]:
        
        def postorder_Rec_gene:
            if node:
                yield from postorder(node.left)
                yield from postorder(node.right)
                yield node.val
                
            return list(postorder(root))
                
        def postorder_Recursion(root: TreeNode):
            if not root: return []
            return self.postorderTraversal(root.left) + self.postorderTraversal(root.right) + [root.val]

        def postorder_Iteration(root: TreeNode):
            if not root: return []
            ans = []
            stack = []
            current = root
            visited = None
            while stack or current:
                if current:
                    stack.append(current)
                    current = current.left
                    continue
                current = stack[-1]
                if current.right and current.right != visited:
                    current = current.right
                else:
                    visited = current
                    ans.append(current.val)
                    stack.pop()
                
            return ans
        
        def postorder_Morris(root: TreeNode):
            ## reverse order of preorder traverse if interchange left and right children
            ans = []
            current = root
            while current:
                ans.append(current.val)
                if not current.right:
                    current = current.left
                else:
                    LeftMost(current.right).left = current.left
                    current.left, current = None, current.right
            return reversed(ans)
        
        def LeftMost(node:TreeNode) -> TreeNode:
            while node.left:
                node = node.left
            return node
        
        return postorder_Iteration(root)
```
